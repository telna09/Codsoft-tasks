**TITANIC DATASET**

The Titanic dataset from Kaggle is used for data analysis and machine learning. Passenger details include ID, name, age, gender, and survival status (`Survived`). `Pclass` is class, often related to socio-economic status, with 1st, 2nd, or 3rd.

The most significant survival factors are gender and age, but women and children are preferentially treated. 'SibSp' and 'Parch' refer to family on board and have implications for survival chances. 'Fare' is the ticket fare and is linked with class, and 'Cabin' refers to one's cabin location on the ship, which would also be related to survival. 'Embarked' accounts for the boarding port (Cherbourg, Queenstown, or Southampton), which might also be related to class.

```python
plt.figure(figsize=(9, 10))

sns.boxplot(df, x='Age')
plt.show()
```

Graph 4 : This box plot, therefore, indicates the distribution of ages across this data set. The large interquartile range of about 20 to 40 indicates that most people fall within this range. The median lies at around 30. There are some outliers at the higher end, with some values of age going as high as about 80, but all other values fall between 0 and 60.

```python
plt.figure(figsize=(9, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.1f')

plt.show()
```

Graph 5 : This heat map gives the correlation of all the numerical features of the Titanic dataset: colors coded in the strength and direction of each correlation (red for positive, blue for negative). The particularly interesting relationships include a negative correlation between Pclass  and Fare (-0.5), indicating higher-class passengers paid more, and a positive correlation between Fare and Survived (0.3), meaning that the best payers had a larger chance of survival. Further, the correlation between SibSp and Parch is not strong; it is moderate with a correlation coefficient of 0.4, suggesting a traveller with siblings/spouses often accompanied parents/children.
